### Instance-based prediction VS prototype-less cluster-based prediction

```python
# RMSE of instance-based vote prediction
plt.figure(figsize=(12, 6))
plt.plot([1, 3, 5, 7, 9, 15, 20], [6.03, 4.93, 4.66, 4.54, 4.51, 4.44, 4.43], 'b-o', label='hist-euclid')
plt.plot([1, 3, 5, 7, 9, 15, 20], [6.08, 4.92, 4.71, 4.57, 4.52, 4.43, 4.39], 'r-o', label='hist-cosine')
plt.plot([1, 3, 5, 7, 9, 15, 20], [5.36, 4.32, 4.13, 4.07, 3.98, 4.05, 4.07], 'g-o', label='incv3-euclid')
plt.plot([1, 3, 5, 7, 9, 15, 20], [5.6, 4.4, 4.24, 4.14, 4.08, 4.02, 4.07], 'y-o', label='incv3-cosine')
plt.xlim(0, 21)
plt.ylim(3.5, 6.5)
plt.xticks([1, 3, 5, 7, 9, 15, 20])
plt.legend()
plt.title('RMSE of instance-based vote prediction')
plt.show()
```

### Metrics for cluster-based vote prediction

```python
plt.figure(figsize=(16, 8))
plt.plot(k_values_cluster, rmse_cluster['incv3_feats_euclid'], 'b-o', label='incv3-euclid')
plt.plot(k_values_cluster, rmse_cluster['incv3_feats_cosine'], 'r-o', label='incv3-cosine')
plt.plot(k_values_cluster, rmse_cluster['color_hist_euclid'], 'g-o', label='hist-euclid')
plt.plot(k_values_cluster, rmse_cluster['color_hist_cosine'], 'y-o', label='hist-cosine')
plt.xlim(0, 12)
plt.ylim(3.5, 5.0)
plt.xticks(k_values_cluster)
plt.xlabel('Number of k nearest prototypes')
plt.ylabel('RMSE')
plt.legend()
plt.title('RMSE of cluster-based vote prediction')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(k_values_cluster, times_cluster['incv3_feats_euclid'], 'b-o', label='incv3-euclid')
plt.plot(k_values_cluster, times_cluster['incv3_feats_cosine'], 'r-o', label='incv3-cosine')
plt.plot(k_values_cluster, times_cluster['color_hist_euclid'], 'g-o', label='hist-euclid')
plt.plot(k_values_cluster, times_cluster['color_hist_cosine'], 'y-o', label='hist-cosine')
plt.xlim(0, 12)
plt.ylim(0, 2000)
plt.xticks(k_values_cluster)
plt.xlabel('Number of k nearest prototypes')
plt.ylabel('Microseconds (us)')
plt.legend()
plt.title('Avg. query time for cluster-based vote prediction')
plt.show()
```
