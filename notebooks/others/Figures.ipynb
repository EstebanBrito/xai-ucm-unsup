```python
import matplotlib.pyplot as plt
```

# Cluster-based prediction VS instance-based prediction

```python
k_values_cluster = [1, 3, 5, 7, 10]
rmse_cluster = {
    'incv3_feats_euclid': [4.11, 4.28, 4.47, 4.48, 4.44],
    'incv3_feats_cosine': [4.05, 4.3, 4.42, 4.46, 4.44],
    'color_hist_euclid': [4.42, 4.51, 4.51, 4.51, 4.51],
    'color_hist_cosine': [4.47, 4.61, 4.61, 4.61, 4.61,]
}
times_cluster = {
    'incv3_feats_euclid': [1613, 1256, 1014, 1054, 1298],
    'incv3_feats_cosine': [1091, 1081, 1187, 1288, 1519],
    'color_hist_euclid': [445, 484, 441, 468, 419],
    'color_hist_cosine': [478, 450, 463, 463, 474]
}
k_values_knn = [1, 3, 5, 10]
rmse_knn = {
    'incv3_feats_euclid': [5.26, 4.31, 4.14, 4.09],
    'incv3_feats_cosine': [5.58, 4.36, 4.26, 4.07],
    'color_hist_euclid': [6.3, 4.99, 4.64, 4.55],
    'color_hist_cosine': [6.39, 5.04, 4.7, 4.48]
}
times_knn = {
    'incv3_feats_euclid': [8915, 8764, 8976, 9000],
    'incv3_feats_cosine': [11178, 11529, 11677, 11669],
    'color_hist_euclid': [3783, 3957, 3938, 4337],
    'color_hist_cosine': [6934, 6944, 7056, 7368]
}
cl_leg = '(usando grupos)'
knn_leg = '(usando instancias)'
```

```python
title_dic = {
    'fontsize': 20,
    'fontweight': 'bold'
}
axis_dic = {
    'fontsize': 18,
    'fontweight': 'normal'
}
tick_dic = {
    'fontsize': 12,
    'fontweight': 'bold'
}
```

## Prediction Performance (RMSE)

```python
plt.figure(figsize=(16, 10))
plt.plot(k_values_cluster, rmse_cluster['incv3_feats_euclid'], 'b.-', label=f'CL-EUC {cl_leg}')
plt.plot(k_values_cluster, rmse_cluster['incv3_feats_cosine'], 'r.-', label=f'CL-COS {cl_leg}')
plt.plot(k_values_cluster, rmse_cluster['color_hist_euclid'], 'g.-', label=f'HIST-EUC {cl_leg}')
plt.plot(k_values_cluster, rmse_cluster['color_hist_cosine'], 'y.-', label=f'HIST-COS {cl_leg}')
plt.plot(k_values_knn, rmse_knn['incv3_feats_euclid'], 'b.--', label=f'CL-EUC {knn_leg}')
plt.plot(k_values_knn, rmse_knn['incv3_feats_cosine'], 'r.--', label=f'CL-COS {knn_leg}')
plt.plot(k_values_knn, rmse_knn['color_hist_euclid'], 'g.--', label=f'HIST-EUC {knn_leg}')
plt.plot(k_values_knn, rmse_knn['color_hist_cosine'], 'y.--', label=f'HIST-COS {knn_leg}')
plt.xlim(0.5, 10.5)
plt.ylim(3.5, 6.5)
plt.xticks(k_values_cluster, **tick_dic)
plt.yticks([4, 5, 6], **tick_dic)
plt.xlabel('Nmero de k casos ms cercanos usados en las prediccin', axis_dic)
plt.ylabel('RMSE', axis_dic)
plt.legend(loc='upper right', fontsize='x-large')
plt.title('RMSE de las predicciones de votos', title_dic)
plt.grid(True)
plt.show()
```

## Time Performance (microseconds)

```python
plt.figure(figsize=(16, 10))
plt.plot(k_values_cluster, times_cluster['incv3_feats_euclid'], 'b.-', label=f'CL-EUC {cl_leg}')
plt.plot(k_values_cluster, times_cluster['incv3_feats_cosine'], 'r.-', label=f'CL-COS {cl_leg}')
plt.plot(k_values_cluster, times_cluster['color_hist_euclid'], 'g.-', label=f'HIST-EUC {cl_leg}')
plt.plot(k_values_cluster, times_cluster['color_hist_cosine'], 'y.-', label=f'HIST-COS {cl_leg}')
plt.plot(k_values_knn, times_knn['incv3_feats_euclid'], 'b.--', label=f'CL-EUC {knn_leg}')
plt.plot(k_values_knn, times_knn['incv3_feats_cosine'], 'r.--', label=f'CL-COS {knn_leg}')
plt.plot(k_values_knn, times_knn['color_hist_euclid'], 'g.--', label=f'HIST-EUC {knn_leg}')
plt.plot(k_values_knn, times_knn['color_hist_cosine'], 'y.--', label=f'HIST-COS {knn_leg}')
plt.xlim(0.5, 13)
plt.ylim(0, 12000)
plt.xticks(k_values_cluster, **tick_dic)
plt.yticks(**tick_dic)
plt.xlabel('Nmero de k casos ms cercanos usados en las prediccin', axis_dic)
plt.ylabel('Microsegundos', axis_dic)
plt.legend(loc='lower right', fontsize='x-large')
plt.title('Tiempo promedio de prediccin de votos', title_dic)
plt.grid(True)
plt.show()
```
