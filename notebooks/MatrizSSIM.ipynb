```python
import numpy as np
import os
from skimage.io import imread, imshow
from skimage.metrics import structural_similarity as ssim
```

```python
import time
```

Load paths for every image in /imgs folder

```python
img_paths = [os.path.join('..', 'imgs', img_name) for img_name in os.listdir('..\\imgs') if img_name.split('.')[-1]=='jpg']
img_paths[:10]
```

```python
len(img_paths)
```

Load the pixels values for every image

```python
values = [imread(img_path) for img_path in img_paths]
```

```python
length = len(img_paths)
sim_matrix = np.zeros(shape=(length, length))
```

## Calculating SSIM matrix
#### TODO: Sim_score = 1 - SSIM (more is less)

```python
def calc_ssim_mx():
    for i in range(length):
        if (i+1)%10==0: print(i+1) # Print every 10 images calculated
        for j in range(i, length):
            # img_1, img_2 = imread(img_paths[i]), imread(img_paths[j])
            img_1, img_2 = values[i], values[j]
            if i==j: sim_score = 1.0
            else: sim_score = ssim(img_1, img_2, window_size=11, multichannel=True)
            # Invert sim score (in clustering, less is more)
            sim_score = 1.0 - sim_score
            sim_matrix[i][j] = sim_score
            sim_matrix[j][i] = sim_score
    return sim_matrix
```

```python
start = time.time()
sim_matrix = calc_ssim_mx()
end = time.time()
print(f'{(start-end)//60} minutes')
```
